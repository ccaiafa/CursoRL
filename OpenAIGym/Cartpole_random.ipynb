5- Sample random action

```python
env.action_space.sample()
```

6- Sample random observation

```python
env.observation_space.sample()
```

7- Implement a loop with random actions

```python
total_reward = 0.0
total_steps = 0
obs = env.reset()

while True:
    action = env.action_space.sample() # take a random action
    obs, reward, done, _ = env.step(action) # execute one step
    total_reward += reward
    total_steps += 1
    if done:
        break

print("Episode done in %d steps, total reward %.2f" % (
    total_steps, total_reward))
```
