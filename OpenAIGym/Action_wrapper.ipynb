**OpenAI Gym: Wrappers**
To modify actions such that 10% of the time a random action is taken

1- Import needed packages

```python
import gym
from typing import TypeVar
import random

!pip install "numpy<2" --quiet
import numpy as np
print("NumPy version:", np.__version__)  # should be 1.26.x

Action = TypeVar('Action')
```

2- Initialize the wrapper by calling a parent's __init__ and saving epsilon. Then, we modify the parent's class action as follows:

```python
class RandomActionWrapper(gym.ActionWrapper):
    def __init__(self, env, epsilon=0.1):
        super(RandomActionWrapper, self).__init__(env)
        self.epsilon = epsilon

    def action(self, action: Action) -> Action:
        if random.random() < self.epsilon:
            print("Random!")
            return self.env.action_space.sample()
        return action
```

3- We apply the action wrapper to the Carpole environment

```python
env = RandomActionWrapper(gym.make("CartPole-v1"))
```

4- We test the wrapper

```python
obs = env.reset()
total_reward = 0.0

while True:
    obs, reward, done, _ = env.step(0)
    total_reward += reward
    if done:
        break

print("Reward got: %.2f" % total_reward)
```
